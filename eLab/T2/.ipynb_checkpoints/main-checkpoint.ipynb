```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import sympy as sp

sin = np.sin
cos = np.cos
pi = np.pi

mpl.rcParams['figure.dpi'] = 300
plt.rcParams.update({"font.family": "Times New Roman", "mathtext.fontset": "cm"})
```

## Problema 1

```python
# a)
R = 1E3 # Ohm
V = 12 # V
Q0 = 3.6E-6 # C
C = 1E-6 # F
alpha = 1/(R*C) # s^-1

t = np.linspace(0, 4/alpha, 1000)
I = (V - Q0/C)/R * np.exp(-alpha * t) * 1000 # mA

fig, ax = plt.subplots()

ax.plot(t, I, label="$I(t)$")
ax.axhline(I[0], label=f"$I_0$ = {I[0]} mA", ls="--", c="#999")
ax.axvline(1/alpha, label=f"$\\tau = RC$ = {1000/alpha} ms", ls=":", c="#999")
ax.set(title="Corriente en el tiempo - Circuito 1", xlabel="Tiempo [s]", ylabel="Corriente [mA]")
plt.legend()

plt.savefig("1-a.png")
plt.show()
```

```python
# b)
R = 1E3 # Ohm
V = 12 # V
L = 1E-3 # H
beta = R/L # s^-1

t = np.linspace(0, 4/beta, 1000)
I = V/R * (1 + np.exp(-beta*t)) * 1000 # mA

fig, ax = plt.subplots()

ax.plot(t, I, label="$I(t)$", c="r")
ax.axhline(I[0], label=f"$I_0$ = {I[0]} mA", ls="--", c="#999")
ax.axhline(V/R*1000, label=f"$I_{{\\infty}}$ = {V/R*1000} mA", ls="--", c="#999")
ax.axvline(1/beta, label=f"$\\tau = L/R$ = {10**6/beta} $\\mu$s", ls=":", c="#999")
ax.set(title="Corriente en el tiempo - Circuito 2", xlabel="Tiempo [s]", ylabel="Corriente [mA]")
plt.legend()

plt.savefig("1-b.png")
plt.show()
```

## Problema 4

```python
Vin, Vout = sp.symbols('V_in V_out')
w, C1, C2, R1, R2 = sp.symbols('w C1 C2 R1 R2')
I1, I2 = sp.symbols('I1 I2')

ZC1 = 1/(1j*w*C1)
ZC2 = 1/(1j*w*C2)
eq1 = (R1 + ZC1)*I1 + R2*(I1 - I2) - Vin
eq2 = R2*(I2 - I1) + ZC2*I2
eq3 = ZC2*I2 - Vout

solutions = sp.solve([eq1, eq2, eq3], (Vout, I1, I2), dict=True)
solutions
```

## Problema 5

```python
@np.vectorize
def Fourier_series(t, N, omega, C0, An, Bn):
    series = C0 / 2
    for n in range(1, N):
        series += An(n)*sin(n*omega*t) + Bn(n)*cos(n*omega*t)
    return series
```

```python
A, tau = 1, 1

t = np.linspace(-2*tau, 2*tau, 1000)
f = A * np.abs(sin(pi/tau * t))
fname = "$A |\\sin(\\pi t / \\tau)|$"

N = 6
omega = pi / tau
C0 = 4*A/pi
An = lambda n: 0
Bn = lambda n: 4*A/pi * (1 - n**2)**(-1) if n % 2 == 0 else 0
f_series = Fourier_series(t, N, omega, C0, An, Bn)

fig, ax = plt.subplots()
ax.plot(t, f, label=fname)
ax.plot(t, f_series, label="Serie de Fourier")
ax.set(title=f"Serie de Fourier de {fname} hasta n = {N}", xlabel="Tiempo [$\\tau$]", ylabel="Amplitud [$A$]")
plt.legend()

plt.savefig("5-a.png")
plt.show()
```

```python
A, tau = 1, 1

t = np.linspace(-2*tau, 2*tau, 1000)
f = 2*A/tau * (t%tau) - A
fname = "$2A (t\,\mathrm{mod}\,\\tau) / \\tau - A$"

N = 6
omega = pi / tau
C0 = 0
An = lambda n: -4*A/(n*pi) if n % 2 == 0 else 0
Bn = lambda n: 0
f_series = Fourier_series(t, N, omega, C0, An, Bn)

fig, ax = plt.subplots()
ax.plot(t, f, label=fname)
ax.plot(t, f_series, label="Serie de Fourier")
ax.set(title=f"Serie de Fourier de {fname} hasta n = {N}", xlabel="Tiempo [$\\tau$]", ylabel="Amplitud [$A$]")
plt.legend()

plt.savefig("5-b.png")
plt.show()
```

## Problema 6

```python
Omega = 1
t = np.linspace(-30/Omega, 30/Omega, 1000)
delta = sin(Omega*t)/(pi*t)

fig, ax = plt.subplots()

ax.plot(t, delta, label="$\\delta_{\\Omega}(t)$", color="r")
ax.vlines(0, ymin=0, ymax=1/pi+0.05, label="$\\delta(t)$")
ax.axhline(1/pi, c="#999", ls="--", label="$\\pi^{-1}$")
ax.axhline(0, c="#999", ls=":")
ax.set(title="Delta de Dirac tras un filtro pasa-bajos ideal", xlabel="Tiempo [$\\Omega^{-1}$]",
      ylabel="Amplitud [u.a.]")
ax.set(ylim=(-0.08, 1/pi+0.05))
plt.legend()

plt.savefig("6.png")
plt.show()
```
