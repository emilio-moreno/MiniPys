```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({"font.family": "Times New Roman", "mathtext.fontset": "cm"})
```

## Problema 1

```python
# a)
R = 1E3 # Ohm
V = 12 # V
Q0 = 3.6E-6 # C
C = 1E-6 # F
alpha = 1/(R*C) # s^-1

t = np.linspace(0, 4/alpha, 1000)
I = (V - Q0/C)/R * np.exp(-alpha * t) * 1000 # mA

fig, ax = plt.subplots()

ax.plot(t, I, label="$I(t)$")
ax.axhline(I[0], label=f"$I_0$ = {I[0]} mA", ls="--", c="#999")
ax.axvline(1/alpha, label=f"$\\tau = RC$ = {1000/alpha} ms", ls=":", c="#999")
ax.set(title="Corriente en el tiempo - Circuito 1", xlabel="Tiempo [s]", ylabel="Corriente [mA]")
plt.legend()

plt.show()
```

```python
# b)
R = 1E3 # Ohm
V = 12 # V
L = 1E-3 # H
beta = R/L # s^-1

t = np.linspace(0, 4/beta, 1000)
I = V/R * (1 + np.exp(-beta*t)) * 1000 # mA

fig, ax = plt.subplots()

ax.plot(t, I, label="$I(t)$", c="r")
ax.axhline(I[0], label=f"$I_0$ = {I[0]} mA", ls="--", c="#999")
ax.axhline(V/R*1000, label=f"$I_{{\\infty}}$ = {V/R*1000} mA", ls="--", c="#999")
ax.axvline(1/beta, label=f"$\\tau = L/R$ = {10**6/beta} $\\mu$s", ls=":", c="#999")
ax.set(title="Corriente en el tiempo - Circuito 2", xlabel="Tiempo [s]", ylabel="Corriente [mA]")
plt.legend()

plt.show()
```
