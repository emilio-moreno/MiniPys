```python
import numpy as np
import matplotlib.pyplot as plt
from typing import List
sin, cos, pi = np.sin, np.cos, np.pi
mu_0 = 1


plt.rcParams.update({"font.family": "Times New Roman", "mathtext.fontset": "cm",
                     "font.size": 8})
plt.style.use("dark_background")
```

```python
def outside_field2d(X: np.ndarray, Y: np.ndarray,
           m: List[float] = [0, 1], normal: bool = False):
    '''Returns an (optionally) unscaled dipole field.'''
    A = mu_0 / 4 * np.pi
    r_dot_m = m[0] * X + p[1] * Y
    r = np.sqrt(X**2 + Y**2)
    
    Bx = (A / r**3) * ((3 * r_dot_m * X) - m[0])
    By = (A / r**3) * ((3 * r_dot_m * Y) - m[1])
    B = np.sqrt(Bx**2 + By**2)
    
    if normal:
        Bx, By = Bx / B, By / B
    
    return Bx, By, B

def inside_field2d(X, Y, M: np.ndarray = np.array([0, 1]), normal: bool = False):
    B0 = 2 * mu_0 * M / 3
    B = np.sqrt(B0[0]**2 + B0[1]**2)
    if normal:
        B0 = B0 / B
    return B0[0] * np.ones(X.shape), B0[1] * np.ones(Y.shape), B

def magnetized_sphere2d(X, Y, r_sphere, m: np.ndarray = np.array([0, 1]), normal = False):
    m = (4 * pi * r_sphere**3) * M / 3
    Bx, By, B = np.zeros(X.shape), np.zeros(Y.shape), np.zeros(X.shape)
    sphere_mask = X**2 + Y**2 < r_sphere**2
    Bx[sphere_mask], By[sphere_mask], B[sphere_mask] = inside_field2d(X[sphere_mask], Y[sphere_mask], m, normal)
    Bx[~sphere_mask], By[~sphere_mask], B[~sphere_mask] = outside_field2d(X[~sphere_mask], Y[~sphere_mask], M, normal)
    return Bx, By, B
```

```python
step = 0.08
length = 2
r_sphere = 0.65

M = np.array((0, 1))
X, Y = np.mgrid[-length/2:length/2:step, -length/2:length/2:step]
p = [0, 1]
Bx, By, B = magnetized_sphere2d(X, Y, r_sphere, M, normal=True)

fig, ax = plt.subplots(dpi=200)
title = "Campo magntico de una esfera magnetizada de radio R = 0.65 L"
ax.set(xlabel="Posicin en $\\rho$ [$L$]", ylabel="Poscin en $z$ [$L$]", title=title)

quiver = ax.quiver(X, Y, Bx, By, B, pivot='mid', cmap=plt.cm.BuPu_r)
x = np.linspace(-r_sphere, r_sphere, 10**3)
ax.plot(x, np.sqrt(r_sphere**2 - x**2), color="white")
ax.plot(x, -np.sqrt(r_sphere**2 - x**2), color="white")
fig.colorbar(quiver, label="Magnitud de campo magntico ($B$) [$\mu_0 M$]",
                 fraction=0.046, pad=0.04)

plt.savefig("Figures/magnetized_sphere.png")
```
