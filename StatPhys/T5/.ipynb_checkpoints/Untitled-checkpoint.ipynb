```python
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams.update({"font.family": "Times New Roman", "mathtext.fontset": "cm"})
```

## Ejercicio 1

```python
T_star = np.arange(3, 5.5, 0.5)
x = np.linspace(-0.999, 0.999, 1000)
U_star = -2*x**2

fig, ax = plt.subplots(dpi=300)
ax.set(title="Energa libre de Helmholtz\n para un sistema de dipolos en la aproximacin de campo medio", xlabel="$x$", ylabel="$F^*$")
for t_star in T_star:
    F_star = U_star + t_star * ((1 + x)/2 * np.log((1 + x)/2) + (1 - x)/2 * np.log((1 - x)/2))
    ax.plot(x, F_star, label=f"$T^{{*}}$ = {t_star}")
plt.legend()

plt.savefig("figures/Ejercicio1.png")
```

## Ejercicio 2

```python
fig, axs = plt.subplot_mosaic([[6, 4, 3]], sharey=True, dpi=300)

x = np.linspace(0, 0.999, 1000)
T_star = [6, 4, 3]
RHS = np.log((1 + x)/(1 - x))

for t_star in T_star:
    LHS = 8 * x / t_star

    if t_star == 6: ylabel = "Ecuacin transcendental ($f(x)$)"
    else: ylabel = ""
    axs[t_star].set(title=f"Soluciones para $T^* =$ {t_star}", xlabel="$x$", ylabel=ylabel)
    axs[t_star].plot(x, RHS)
    axs[t_star].plot(x, LHS)

plt.savefig("figures/Ejercicio2.png")
```

Para temperaturas mayores o iguales a la temperatura crtica, obtenemos soluciones con $x = 0$, que corresponden a la posicin de los mnimos centrales en las grficas del Ejercicio 1. Por su lado, para temperaturas menores a la temperatura crtica, tenemos una solucin en $x = x_0 \neq 0$, lo que produce dos mnimos en $x = \pm x_0$, que se manifiestan en este rgimen de temperaturas para las grficas del Ejercicio 1.

Observamos que el sistema tiende a saturarse para temperaturas altas ($T^* > 2.5$), pues el aumento de temperatura produce cambios cada vez menores en la energa interna, llevando a la reduccin de la capacidad calorfica al pasar su mximo en $T^* \approx 0.5$. Por otro lado, para el rgimen de temperatura bajas ($T^* < 0.25$), la energa y la capacidad calorfica incrementan rpidamente con la temperatura.
